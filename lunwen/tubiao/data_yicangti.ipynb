```python
filePath = r'E:\tools\写论文\datas\data\yicangti\3'
name = os.listdir(filePath)
```

```python
vps = []
for model in name:
    if model.endswith('vp'):
        vps.append(model)
```

```python
len(vps)
```

```python
import random
```

```python
random.shuffle(vps)
vpdata1 = np.fromfile(r'E:\tools\写论文\datas\data\yicangti\3\\'+vps[0],dtype=np.float32)
vpdata1 = vpdata1.reshape((512,512))
vpdata2 = np.fromfile(r'E:\tools\写论文\datas\data\yicangti\3\\'+vps[1],dtype=np.float32)
vpdata2 = vpdata2.reshape((512,512))
vpdata3 = np.fromfile(r'E:\tools\写论文\datas\data\yicangti\3\\'+vps[2],dtype=np.float32)
vpdata3 = vpdata3.reshape((512,512))
```

```python
plt.figure(figsize=(18,12))
plt.subplot(1, 3, 1)
plt.title("数据1")
plt.ylabel("深度")
plt.xlabel("水平")
plt.imshow(vpdata1.T)
plt.subplot(1, 3, 2)
plt.title("数据2")
plt.ylabel("深度")
plt.xlabel("水平")
plt.imshow(vpdata2.T)
plt.subplot(1, 3, 3)
plt.title("数据3")
plt.ylabel("深度")
plt.xlabel("水平")
plt.imshow(vpdata3.T)
plt.savefig(r'E:\tools\写论文\datas\data\yicangti\异常体复杂模型2图片.jpg',bbox_inches='tight',dpi=300,pad_inches=0.5)
```

```python
vpdatas = [vpdata1.T,vpdata2.T,vpdata3.T]
Nr = 1
Nc = 3

fig, axs = plt.subplots(Nr, Nc,figsize=(12, 8))
fig.suptitle('复杂地层下异常体模型',y=0.75,fontsize='xx-large')

images = []
# for i in range(Nr):
for j in range(Nc):
    # Generate data with a range that varies from one plot to the next.
    images.append(axs[j].imshow(vpdatas[j]))
    axs[j].label_outer()

# Find the min and max of all colors for use in setting the color scale.
vmin = min(image.get_array().min() for image in images)
vmax = max(image.get_array().max() for image in images)
norm = colors.Normalize(vmin=vmin, vmax=vmax)
for im in images:
    im.set_norm(norm)
    
position=fig.add_axes([0.13, 0.22, 0.765, 0.04])#位置[左,y,右,高]
cbar  = fig.colorbar(images[0], ax=axs, cax=position, orientation='horizontal', fraction=.2)
cbar.set_label('速度',fontsize='xx-large')

# Make images respond to changes in the norm of other images (e.g. via the
# "edit axis, curves and images parameters" GUI on Qt), but be careful not to
# recurse infinitely!
def update(changed_image):
    for im in images:
        if (changed_image.get_cmap() != im.get_cmap()
                or changed_image.get_clim() != im.get_clim()):
            im.set_cmap(changed_image.get_cmap())
            im.set_clim(changed_image.get_clim())


for im in images:
    im.callbacksSM.connect('changed', update)
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.savefig(r'E:\tools\写论文\datas\data\yicangti\异常体复杂模型2图片.jpg',bbox_inches='tight',dpi=300,pad_inches=0.2)
plt.show()
```
