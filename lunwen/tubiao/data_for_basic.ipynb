```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
x = [i for i in range(-50,50)]
x = np.array(x)
y = np.where(x>0,x,0)
plt.title("y = relu(x)",y=1.2)
# plt.ylabel("y")
# plt.xlabel("x")
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
# ax.xaxis.set_label_position('top')
# ax.yaxis.set_label_position('right')
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.legend(['relu(x)'])
plt.plot(x,y)
plt.savefig(r'E:\tools\写论文\图表数据\basic\relu函数.jpg',bbox_inches='tight',dpi=300,pad_inches=0.5)
```

```python
x = np.linspace(-10,10,100000)
y = 1/(1+np.exp(-x))
plt.title("y = sigmoid(x)",y=1.2)
# plt.ylabel("y")
# plt.xlabel("x")
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
# ax.xaxis.set_label_position('top')
# ax.yaxis.set_label_position('right')
plt.xlim(-10, 10)
plt.ylim(-2, 2)
plt.legend(['sigmoid(x)'])
plt.plot(x,y)
plt.savefig(r'E:\tools\写论文\图表数据\basic\sigmoid函数.jpg',bbox_inches='tight',dpi=300,pad_inches=0.5)
```

```python
x = [i for i in range(-50,50)]
x = np.array(x)
y = np.where(x>0,x,0.2*x)
plt.title("y = leakyRelu(x)",y=1.2)
# plt.ylabel("y")
# plt.xlabel("x")
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
# ax.xaxis.set_label_position('top')
# ax.yaxis.set_label_position('right')
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.legend(['relu(x)'])
plt.plot(x,y)
plt.savefig(r'E:\tools\写论文\图表数据\basic\LeakyReLU函数.jpg',bbox_inches='tight',dpi=300,pad_inches=0.5)
```

```python
x = np.linspace(-10,10,100000)
y = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
plt.title("y = tanh(x)",y=1.2)
# plt.ylabel("y")
# plt.xlabel("x")
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
# ax.xaxis.set_label_position('top')
# ax.yaxis.set_label_position('right')
plt.xlim(-10, 10)
plt.ylim(-2, 2)
plt.legend(['relu(x)'])
plt.plot(x,y)
plt.savefig(r'E:\tools\写论文\图表数据\basic\tanh函数.jpg',bbox_inches='tight',dpi=300,pad_inches=0.5)
```
